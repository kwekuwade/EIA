```python
#Weekly Working Gas in Underground Storage

import requests, json

#Set EAI API
eai_api = "<EIA API>" #Register for an API: https://www.eia.gov/opendata/register.php

#Natural Gas Storage Category Id
category_id = 1709237

api_category_url = "http://api.eia.gov/category/?api_key={}&category_id={}"
api_series_url = "http://api.eia.gov/series/?api_key={}&series_id={}"

#Natural Gas Series Ids:
#Weekly Nonsalt Region Natural Gas Working Underground Storage, Weekly
salt_region = "NG.NW2_EPG0_SNO_R33_BCF.W"

#Weekly East Region Natural Gas Working Underground Storage, Weekly
east_region = "NG.NW2_EPG0_SWO_R31_BCF.W"

#Weekly Midwest Region Natural Gas Working Underground Storage, Weekly
midwest_region = "NG.NW2_EPG0_SWO_R32_BCF.W"

#Weekly South Central  Region Natural Gas Working Underground Storage, Weekly
south_central_region = "NG.NW2_EPG0_SWO_R33_BCF.W"

#Weekly Mountain Region Natural Gas Working Underground Storage, Weekly
mountain_region = "NG.NW2_EPG0_SWO_R34_BCF.W"

#Weekly Pacific Region Natural Gas Working Underground Storage, Weekly
pacific_region = "NG.NW2_EPG0_SWO_R35_BCF.W"

#Weekly Lower 48 States Natural Gas Working Underground Storage, Weekly
lower_48 = "NG.NW2_EPG0_SWO_R48_BCF.W"

#Get most current data
full_report_request = requests.get(api_category_url.format(eai_api, category_id))

if full_report_request.status_code == 200:
    print("Full Report:")
    print(full_report_request.json())

lower_48_series_request = requests.get(api_series_url.format(eai_api, lower_48))

if lower_48_series_request.status_code == 200:
    print("\nSeries Request:")
    print(lower_48_series_request.json())
```
